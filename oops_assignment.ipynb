1. What is Object-Oriented Programming (OOP)?

3. What is an object in OOP?

An object is an instance of a class. It contains data and methods that can be used to perform specific tasks.

Abstraction hides unnecessary details and shows only important features to the user. Encapsulation hides data by restricting direct access and allows access through methods.

5. What are dunder methods in Python?

6. Explain the concept of inheritance in OOP.

Inheritance allows one class to acquire the properties and methods of another class. It promotes code reusability and reduces duplication.

7. What is polymorphism in OOP?

Polymorphism means "many forms." It allows the same method name to perform different actions depending on the object that calls it.

8. How is encapsulation achieved in Python?

Encapsulation is achieved by making attributes private using double underscores (__) and providing methods to access or modify them.

9. What is a constructor in Python?

A constructor is a special method called **init**(). It is automatically executed when an object is created and is used to initialize object attributes.

10. What are class and static methods in Python?

A class method works with class-level data and uses the @classmethod decorator. A static method does not depend on class or instance data and uses the @staticmethod decorator.

11. What is method overloading in Python?

Method overloading means having multiple methods with the same name but different parameters. Python does not support it directly, but similar behavior can be achieved using default arguments.

12. What is method overriding in OOP?

Method overriding occurs when a child class provides its own implementation of a method already defined in the parent class.

13. What is a property decorator in Python?

The @property decorator allows a method to be accessed like an attribute. It is commonly used for controlled access to class data.

14. Why is polymorphism important in OOP?

Polymorphism improves flexibility and code reusability. It allows different objects to be treated in a similar way while performing different actions.

15. What is an abstract class in Python?

An abstract class is a class that cannot be instantiated directly. It serves as a blueprint for other classes and may contain abstract methods.

16. What are the advantages of OOP?

The main advantages of OOP are code reusability, modularity, easier maintenance, better security through encapsulation, and flexibility through inheritance and polymorphism.

17. What is the difference between a class variable and an instance variable?

A class variable is shared by all objects of a class, whereas an instance variable belongs to a specific object and can have different values for different objects.

18. What is multiple inheritance in Python?

Multiple inheritance is a feature where a class inherits properties and methods from more than one parent class.

19. Explain the purpose of **str** and **repr** methods in Python.

The **str** method returns a user-friendly string representation of an object. The **repr** method returns a detailed representation mainly used for debugging.

20. What is the significance of the super() function in Python?

The super() function is used to access methods and constructors of the parent class. It helps in code reuse and simplifies inheritance.

21. What is the significance of the **del** method in Python?

The **del** method is called when an object is about to be destroyed. It is used for cleanup tasks such as releasing resources.

22. What is the difference between @staticmethod and @classmethod in Python?

23. How does polymorphism work in Python with inheritance?

**Question:** Create a `Circle` class with a private attribute `_radius`. Implement a property `radius` that allows setting and getting the radius, ensuring the radius is always a positive number. Also, add a method `area()`.

```python
import math

class Circle:
    def __init__(self, radius):
        self._radius = 0 # Initialize with a default value
        self.radius = radius # Use the setter to validate

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value > 0:
            self._radius = value
        else:
            print("Radius must be positive.")

    def area(self):
        return math.pi * (self.radius ** 2)

# Create a Circle object
my_circle = Circle(5)
print(f"Initial Radius: {my_circle.radius}")
print(f"Area: {my_circle.area():.2f}")

# Try setting a new valid radius
my_circle.radius = 7
print(f"New Radius: {my_circle.radius}")
print(f"New Area: {my_circle.area():.2f}")

# Try setting an invalid radius
my_circle.radius = -2
print(f"Radius after invalid attempt: {my_circle.radius}")
```

**Question:** Create a `Car` class. Add a class variable `number_of_cars` to track how many car objects are created. Implement a `classmethod` to create a `Car` object from a string (e.g., "ModelX-Tesla-2023") and a `staticmethod` to validate if a given year is a vintage year (e.g., before 1990).

```python
class Car:
    number_of_cars = 0 # Class variable

    def __init__(self, model, brand, year):
        self.model = model
        self.brand = brand
        self.year = year
        Car.number_of_cars += 1 # Increment class variable on object creation

    def display_info(self):
        print(f"Model: {self.model}, Brand: {self.brand}, Year: {self.year}")

    @classmethod
    def from_string(cls, car_string):
        model, brand, year = car_string.split('-')
        return cls(model, brand, int(year))

    @staticmethod
    def is_vintage(year):
        return year < 1990

# Create car objects
car1 = Car("Civic", "Honda", 2020)
car2 = Car.from_string("ModelS-Tesla-2018")

car1.display_info()
car2.display_info()

print(f"Total number of cars: {Car.number_of_cars}")

# Use static method
print(f"Is car1 vintage (2020)? {Car.is_vintage(car1.year)}")
print(f"Is a 1985 car vintage? {Car.is_vintage(1985)}")
```

**Question:** Create a `Student` class and a `Course` class. Implement a many-to-many relationship where a student can enroll in multiple courses and a course can have multiple students. Show how to enroll students in courses.

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.courses = [] # List of Course objects

    def enroll_in_course(self, course):
        if course not in self.courses:
            self.courses.append(course)
            course.add_student(self)

    def __str__(self):
        return f"Student: {self.name} (ID: {self.student_id})"

class Course:
    def __init__(self, name, course_code):
        self.name = name
        self.course_code = course_code
        self.students = [] # List of Student objects

    def add_student(self, student):
        if student not in self.students:
            self.students.append(student)

    def __str__(self):
        return f"Course: {self.name} (Code: {self.course_code})"

# Create students
s1 = Student("Alice", "S001")
s2 = Student("Bob", "S002")
s3 = Student("Charlie", "S003")

# Create courses
c1 = Course("Mathematics", "MATH101")
c2 = Course("Physics", "PHY202")
c3 = Course("Chemistry", "CHEM303")

# Enroll students in courses
s1.enroll_in_course(c1)
s1.enroll_in_course(c2)
s2.enroll_in_course(c1)
s3.enroll_in_course(c3)

# Demonstrate enrollments
print(f"{s1.name} is enrolled in:")
for course in s1.courses:
    print(f"  - {course.name}")

print(f"\nStudents in {c1.name}:")
for student in c1.students:
    print(f"  - {student.name}")
```

**Question:** Create a `BankAccount` class with a `balance` attribute. Implement methods for `deposit`, `withdraw`, and handling insufficient funds. Use the `__del__` method to print a message when an account is closed.

```python
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.account_number = account_number
        self.balance = initial_balance
        print(f"Account {self.account_number} created with balance: ${self.balance}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.balance}")
            else:
                print(f"Insufficient funds. Current balance: ${self.balance}")
        else:
            print("Withdrawal amount must be positive.")

    def __del__(self):
        print(f"Account {self.account_number} has been closed.")

# Create an account
account1 = BankAccount("12345", 100)
account1.deposit(50)
account1.withdraw(30)
account1.withdraw(200) # Insufficient funds

# Delete the account to see __del__ in action
del account1
```

**Question:** Demonstrate method chaining using a simple `Calculator` class with `add`, `subtract`, and `multiply` methods. Each method should return `self` to allow chaining.

```python
class Calculator:
    def __init__(self, initial_value=0):
        self.result = initial_value

    def add(self, num):
        self.result += num
        return self # Return self for chaining

    def subtract(self, num):
        self.result -= num
        return self # Return self for chaining

    def multiply(self, num):
        self.result *= num
        return self # Return self for chaining

    def get_result(self):
        return self.result

# Demonstrate method chaining
calc = Calculator(10)
final_result = calc.add(5).subtract(2).multiply(3).get_result()
print(f"Final Calculation Result: {final_result}") # (10 + 5 - 2) * 3 = 39
```

**Question:** Create a `Point` class with `x` and `y` coordinates. Overload the `+` operator so that adding two `Point` objects results in a new `Point` object with summed coordinates.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Point):
            return Point(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Can only add a Point object to another Point object.")

    def __str__(self):
        return f"Point({self.x}, {self.y})"

# Create Point objects
p1 = Point(1, 2)
p2 = Point(3, 4)

# Add points using the overloaded '+' operator
p3 = p1 + p2

print(f"Point 1: {p1}")
print(f"Point 2: {p2}")
print(f"Point 1 + Point 2: {p3}") # Expected: Point(4, 6)
```

**Question:** Implement a `Logger` class using the Singleton design pattern. Ensure that only one instance of `Logger` can exist throughout the application.

```python
class Logger:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(Logger, cls).__new__(cls)
            cls._instance.log_messages = []
            print("Logger created (first instance).")
        else:
            print("Logger already exists (returning existing instance).")
        return cls._instance

    def log(self, message):
        self.log_messages.append(message)
        print(f"LOG: {message}")

# Get two instances of Logger
logger1 = Logger()
logger2 = Logger()

# Check if they are the same instance
print(f"logger1 is logger2: {logger1 is logger2}")

# Log messages using both references
logger1.log("Application started.")
logger2.log("User logged in.")

# Verify that both references share the same log messages
print(f"All log messages: {logger1.log_messages}")
```

**Question:** Create a `Temperature` class that can store temperature in Celsius. Implement conversion to Fahrenheit and Kelvin using properties or methods.

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        self._celsius = value

    @property
    def fahrenheit(self):
        return (self.celsius * 9/5) + 32

    @property
    def kelvin(self):
        return self.celsius + 273.15

# Create a Temperature object
temp = Temperature(25)
print(f"Celsius: {temp.celsius}°C")
print(f"Fahrenheit: {temp.fahrenheit}°F")
print(f"Kelvin: {temp.kelvin}K")

# Change Celsius and see conversions update
temp.celsius = 0
print(f"\nCelsius: {temp.celsius}°C")
print(f"Fahrenheit: {temp.fahrenheit}°F")
print(f"Kelvin: {temp.kelvin}K")
```

**Question:** Create a `Fraction` class that stores numerator and denominator. Implement methods to add and multiply fractions. Ensure fractions are simplified.

```python
import math

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator
        self._simplify()

    def _simplify(self):
        gcd_val = math.gcd(self.numerator, self.denominator)
        self.numerator //= gcd_val
        self.denominator //= gcd_val
        if self.denominator < 0: # Ensure denominator is positive
            self.numerator *= -1
            self.denominator *= -1

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

    def __add__(self, other):
        new_numerator = (self.numerator * other.denominator) + (other.numerator * self.denominator)
        new_denominator = self.denominator * other.denominator
        return Fraction(new_numerator, new_denominator)

    def __mul__(self, other):
        new_numerator = self.numerator * other.numerator
        new_denominator = self.denominator * other.denominator
        return Fraction(new_numerator, new_denominator)

# Create fractions
f1 = Fraction(1, 2)
f2 = Fraction(3, 4)
f3 = Fraction(2, 6)

print(f"f1: {f1}")
print(f"f2: {f2}")
print(f"f3 (simplified): {f3}")

# Perform operations
sum_f = f1 + f2
product_f = f1 * f2

print(f"f1 + f2 = {sum_f}") # Expected: 5/4
print(f"f1 * f2 = {product_f}") # Expected: 3/8
```

**Question:** Implement a custom iterator `MyRange` that behaves like Python's built-in `range()` function (e.g., `for i in MyRange(5): print(i)`).

```python
class MyRange:
    def __init__(self, start, end=None, step=1):
        if end is None:
            self.current = 0
            self.end = start
        else:
            self.current = start
            self.end = end
        self.step = step

    def __iter__(self):
        return self

    def __next__(self):
        if (self.step > 0 and self.current < self.end) or \
           (self.step < 0 and self.current > self.end):
            value = self.current
            self.current += self.step
            return value
        else:
            raise StopIteration

# Demonstrate MyRange
print("MyRange(5):")
for i in MyRange(5):
    print(i)

print("\nMyRange(2, 8):")
for i in MyRange(2, 8):
    print(i)

print("\nMyRange(10, 0, -2):")
for i in MyRange(10, 0, -2):
    print(i)
```

**Question:** Create a `Counter` class that implements the `__call__` method, allowing an instance of the class to be called like a function to increment and return its count.

```python
class Counter:
    def __init__(self, initial_count=0):
        self.count = initial_count

    def __call__(self):
        self.count += 1
        return self.count

# Create a counter instance
my_counter = Counter()

print(f"Initial count: {my_counter.count}")

# Call the instance like a function
print(f"Count after first call: {my_counter()}") # 1
print(f"Count after second call: {my_counter()}") # 2

# Another instance
another_counter = Counter(10)
print(f"\nAnother counter initial: {another_counter.count}")
print(f"Another counter after call: {another_counter()}") # 11
```
